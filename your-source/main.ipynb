## Challenge 1: Fork Languages

### Obtain the full list of forks created from the main lab repo via Github API

```python
# Function to get the fork information from each page
# headers contain the token for login

def get_fork(headers):
    
    # Parameters to modify for other repos
    user = 'ironhack-datalabs'
    repo = 'mad-oct-2018'
    
    # Page counter
    page = 1
    
    # Output list containing all the information
    output = []
    
    while True:
        
        request = f'https://api.github.com/repos/{user}/{repo}/forks?page={page}'
        r = requests.get(request, headers=headers)
        fork = r.json()

        # check if fork is empty or not
        if len(fork)== 0:
            break
        
        if page == 1:
            # Append the information to the output
            df = pd.json_normalize(fork)
        else:
            # Append the information to the output
            df_aux = pd.json_normalize(fork)
            df = pd.concat([df, df_aux], ignore_index=True)
        
        page += 1
        
    return df
```

## Challenge 2: Count Commits

### Obtain all the commits made in the past week via API, which is a JSON array that contains multiple commit objects

As the repository was updated last time last in 2018. I will obtain the commits since 2018-08-17T00:00:00Z

```python
# Function to get the commit information from each page (in case that there is more than one page)
# headers contain the token for login

def get_commit(headers):
    
    # Parameters to modify for other repos
    user = 'ironhack-datalabs'
    repo = 'mad-oct-2018'
    date = '2018-08-17T00:00:00Z'

    # Page counter
    page = 1

    # Output list containing all the information
    commits = []

    while True:
        
        request = f'https://api.github.com/repos/{user}/{repo}/commits?since={date}&page={page}'
        r = requests.get(request, headers=headers)
        commit = r.json()

        # check if fork is empty or not
        if len(commit)== 0:
            break

        # Append the information to the output
        commits.append(commit)

        page += 1


    return commits
```

## Challenge 3: Hidden Cold Joke

```python
def find_files(headers):
    
    # Parameters to modify for other repos
    user = 'ironhack-datalabs'
    repo = 'scavenger'

    # List containing the files
    files = []

    # Obtain all the directories
    request = f'https://api.github.com/repos/{user}/{repo}/contents?'
    r = requests.get(request, headers=headers).json()
    
    # Loop over all the directories
    for directory in r:
        
        if directory['type'] == 'dir':
            r1  = requests.get(directory['url'], headers=headers).json()
            
            # check if there is a file with its filename containing .scavengerhunt
            for file in r1:
                if 'scavengerhunt' in file['name']:
                    files.append({'name':file['name'], 'url':file['url']})
                    
    # Sort the files (ascending)
    files = sorted(files, key=lambda k: k['name'])
    
    return files
```

```python
def get_files_content(files_list):
    
    # Parameters to modify for other repos
    user = 'ironhack-datalabs'
    repo = 'scavenger'
    
    # list with the content of each file
    content = []
    
    for file in files_list:

        # Get the file information
        r = requests.get(file['url'], headers=headers).json()
        string = r['content']
        #Convert the the content of the file to ascii
        # It is in base64 (information read from the request in the filed encoding)
        string = str(base64.b64decode(string))
        string = string.replace("\\n'", "").replace("b'","")
        content.append(string)
        
    return content
```

```python
# Obtain the list of files in ascending order
files = find_files(headers)

# Obtain the content of the files
content = get_files_content(files)

# Concatenate the strings
joke = ' '.join(content)

# Print the hidden joke
print(joke)
```
